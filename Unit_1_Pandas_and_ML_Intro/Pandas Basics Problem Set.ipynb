12) Read the `NBA_season1718_salary.csv` file into a Pandas dataframe.

Answer the following questions about the data (put your answers in this Markdown Cell)
1) What was the mean and median NBA player salary across all teams in 2017-18? 

2) Who was the second highest paid Blazer in 2017-2018?

3) What team had the highest average salary per player?  What was that average salary? (you may want to read about the groupby method in Pandas)

4) What NBA team had the most expensive total roster in 2017-2018?

```python
# code for 12
import pandas as pd
nba = pd.read_csv('NBA_season1718_salary.csv', index_col = 0)
nba.columns = nba.columns.str.strip()
print(f"mean salary: {nba.season17_18.mean()}")
print(f"median salary: {nba.season17_18.median()}")
print()

sorted_list0 = nba.season17_18.value_counts().sort_index(ascending = False)
print(f"second high paid: {sorted_list0.index[1]}")

temp = nba.groupby("Tm").agg({'season17_18': 'mean'})
print(f"the highest paid team: {temp.index[0]}")
print(f"the highest team's salary: {temp.iloc[0][0]}")
print()

temp0 = nba.groupby("Tm").agg({'season17_18': 'sum'})
print(f"the highest roster team: {temp0.index[0]}")
print(f"the highest team's roster fee: {temp0.iloc[0][0]}")
```
